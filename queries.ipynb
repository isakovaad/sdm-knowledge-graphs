5.Find the most prolific authors in database conferences: this query retrieves authors who have published the highest number of papers in database conferences. This can help identify key contributors in the field.


PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX : <http://example.org/ontology#>  # Adjust this to your actual namespace

SELECT ?author (COUNT(?paper) AS ?paperCount)
WHERE {
    ?paper :writtenBy ?author .
    ?paper :publishedIn ?conference .
    ?conference rdf:type :DatabaseConference .
}
GROUP BY ?author
ORDER BY DESC(?paperCount)
LIMIT 10


6.Find collaboration networks of authors in database conference: this query identifies pairs of authors who have co-authored papers in database conferences, which can be useful for understanding collaboration patterns and networks.

PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX : <http://example.org/ontology#>  # Adjust this to your actual namespace

SELECT DISTINCT ?author1 ?author2
WHERE {
    ?paper :writtenBy ?author1 .
    ?paper :writtenBy ?author2 .
    ?paper :publishedIn ?conference .
    ?conference rdf:type :DatabaseConference .
    FILTER (?author1 != ?author2)
}
